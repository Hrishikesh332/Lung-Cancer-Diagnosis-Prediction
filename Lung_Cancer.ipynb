# Pre-Processing of the Dataset:
The data present in the gender column is in male and female and in the Lung Cancer column the data is present in boolean form. So it is difficult to use this column for visualization purpose.

So, Here we will convert the gender and the Lung cancer column into the int64 data type column.


```python
from sklearn.preprocessing import LabelEncoder

label=LabelEncoder()
df['LUNG_CANCER']=label.fit_transform(df.LUNG_CANCER)
print(df.head())
```

In above result we can  observe that the LUNG_CANCER column contains the label 'YES' and 'NO'. By using the Label Encode we have converted the label into the numeric value as follows:


*   YES - 1
*   NO  - 0



```python
df['GENDER']=label.fit_transform(df.GENDER)
print(df.head())
```

Again, we used the label encoder on the gender column. The gender column contains the label such as 'M' and 'F'. By using the Label Encode we have converted the label into the numeric value as follows:



*   M - 1
*   F - 0

# **Model Analysis of the data:**

```python
df.head()
```

```python
# Splitting the data for training and testing
from sklearn.model_selection import train_test_split
y = df['LUNG_CANCER']
X = df.drop('LUNG_CANCER', axis=1)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# **Random Forest Classifier**

The random forest uses many trees, and it makes a prediction by averaging the predictions of each component tree. It generally has much better predictive accuracy than a single decision tree and it works well with default parameters. If you keep modeling, you can learn more models with even better performance, but many of those are sensitive to getting the right parameters.

```python
# Building a ML model for the data
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=500, random_state=1)
model.fit(x_train, y_train)
prediction = model.predict(x_test)
```

# **What is Model Validation?**

You'll want to evaluate almost every model you ever build. In most (though not all) applications, the relevant measure of model quality is predictive accuracy. In other words, will the model's predictions be close to what actually happens.

Many people make a huge mistake when measuring predictive accuracy. They make predictions with their training data and compare those predictions to the target values in the training data. You'll see the problem with this approach and how to solve it in a moment, but let's think about how we'd do this first.

You'd first need to summarize the model quality into an understandable way. If you compare predicted and actual home values for 10,000 houses, you'll likely find mix of good and bad predictions. Looking through a list of 10,000 predicted and actual values would be pointless. We need to summarize this into a single metric.

# **Mean Absolute Error**

There are many metrics for summarizing model quality, but we'll start with one called Mean Absolute Error (also called MAE). Let's break down this metric starting with the last word, error.

The prediction error for each house is:

error=actualâˆ’predicted
So, if a house cost USD150,000 and you predicted it would cost  USD100,000 the error is USD50,000.

```python
# Checking the validity of the model
from sklearn.metrics import mean_absolute_error as mae
# print(mean absolute error value for the random forest regressor model)
print(mae(y_test, prediction))
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test, prediction))
```

```python
from sklearn.metrics import precision_score
prec=precision_score(y_test, prediction)
print("Precision Score = ",prec)
```

# **Gradient Boosting Classifier:**

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import precision_score
model1=GradientBoostingClassifier(n_estimators= 1000, learning_rate=0.05)
model1.fit(x_train, y_train)
pred=model1.predict(x_test)
prec1=precision_score(y_test, pred)
print("Precision Score = ",prec1)
```

# **Logistic Regression:**

```python
from sklearn.linear_model import LogisticRegression
model2=LogisticRegression()
model2.fit(x_train, y_train)
pred2=model2.predict(x_test)
prec2=precision_score(y_test, pred2)
print('Mean Absolute Error for the logistic regression model:', mae(y_test, pred2))
print("Precision Score = ",prec2)
```

```python
#Confusion matrix for Logistic Regression
from sklearn.metrics import confusion_matrix  
conf= confusion_matrix(y_test, pred2)
print(conf)
```

# **Support Vector Classifier:**

```python
from sklearn.svm import SVC
model3=SVC(random_state=1)
model3.fit(x_train, y_train)
pred3=model3.predict(x_test)
prec3=precision_score(y_test, pred3)
print('Mean Absolute Error for the Support Vector Machine model:', mae(y_test, pred3))
print("Precision Score = ",prec3)
```

```python
models=["Gradient Boosting Classifier","Logistic Regression","Random Forest Classifier","Support Vector Classifier"]
prec_values=[prec1, prec2, prec, prec3]
figure=(px.bar(x=models,y=prec_values,text=prec_values,
            text_auto=True,color=models,height=700,width=850,template="plotly_dark", 
                        title='Comparision of the precision score of different models:'))
figure.show()
```
